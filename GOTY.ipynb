```python
# Visit VG Charts
games = []
#how to get only super mario maker 1, overwatch 1, doom, inside, God of War, Marvels Spider Man 1, Red Dead Redemption 1,
# resident evil 2, The outer Worlds, death stranding, control, Hades, Doom Eternal, Resident Evil 8 Village, Stray
games_to_search = ['Dragon Age: Inquisition','Hearthstone: Heroes of Warcraft',
                   'Darks Souls 2','Bayonetter 2','Middle-Earth: Shadow of Mordor',
                   'The Witcher 3: Wild Hunt', 'Metal Gear solid V: The Phantom Pain', 'Bloodborne',
                   'Super Mario Maker', 'Fallout 4', 'Overwatch', 'Titanfall 2', 'Doom', 'Inside',
                   'Uncharted 4: A Thiefs End', 'Breath of the Wild', 'PlayerUnknown Battlegrounds', 
                   'Horizon Zero Dawn','Super Mario Odyssey', 'Persona 5 Royal', 'God of War', 
                   'Monster Hunter: World', 'Assassin Creed Odyssey', 'Marvels Spider-Man', 'Red Dead Redemption',
                   'Sekiro: Shadows Die Twice', 'Resident Evil 2', 'The Outer Worlds', 'Death Stranding', 'Control', 
                   'Last of Us Part II', 'Hades', 'Animal Crossing: New Horizons', 'Doom Eternal', 'Ghost of Tsushima',
                   'Final Fantasy VII Remake', 'It Takes Two', 'Resident Evil 8 Village', 'Metroid Dread', 'Deathloop',
                   'Ratchet & Clank Rift Apart', 'Psychonauts 2', 'Elden Ring', 'Xenoblade Chronicles 3', 
                   'A Plague Tale Requiem','Horizon Forbidden West', 'Stray', 'Alan Wake 2', 'Baldur Gate III',
                   'Marvel Spider Man 2', 'Resident Evil 4', 'Super Mario Bros Wonder', 
                   'Legend of Zelda Tears of the Kingdom']
for bob in games_to_search: 
    url = 'https://www.vgchartz.com/games/games.php?name=&keyword=&console=&region=All&developer=&publisher=&goty_year=&genre=&boxart=Both&banner=Both&ownership=Both&showmultiplat=No&results=50&order=Sales&showtotalsales=0&showtotalsales=1&showpublisher=0&showpublisher=1&showvgchartzscore=0&shownasales=0&shownasales=1&showdeveloper=0&showdeveloper=1&showcriticscore=0&showcriticscore=1&showpalsales=0&showpalsales=1&showreleasedate=0&showreleasedate=1&showuserscore=0&showuserscore=1&showjapansales=0&showjapansales=1&showlastupdate=0&showlastupdate=1&showothersales=0&showothersales=1&showshipped=0&showshipped=1'
    browser.visit(url)
    browser.find_by_css('input[name="name"]').fill(bob)
    browser.find_by_text('Search')[0].click()
    search_result = browser.find_by_css('#generalBody').find_by_css('tbody')[0]
    #hashtag is id, class is .., tag is just name of tag
    search_result
    table_rows = search_result.find_by_css('tr')[3:]

    for row in table_rows:
        data = []
        for cell in row.find_by_css('td'): 
            cell_text = cell.text
            data.append(cell_text)
        games.append(data)

games
```
